```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df_datos = pd.read_excel("datos.xlsx")
df_datos.head(5)
```

```python
df_datos["precio_unitario"] = df_datos["MONTO FINAL (INC IVA)"]/df_datos["CANTIDAD"]
df_datos
```

```python
print(f"Valores únicos ID: {df_datos['ID'].nunique()}")
print(f"Valores únicos CODIGO: {df_datos['CODIGO'].nunique()}")
print(f"Valores únicos ITEM: {df_datos['ITEM'].nunique()}")
```

```python
df_columnas = df_datos[["ID","CANTIDAD", "MONTO FINAL (INC IVA)", "FACTURA", "FECHA EMISIÓN", "precio_unitario"]]
df_columnas
```

```python
df_cantidades = df_columnas.groupby("ID").sum("CANTIDAD")[["CANTIDAD"]].sort_values("CANTIDAD", ascending=False).reset_index()
df_cantidades
```

```python
#.loc .iloc para filtrar 

IDs_filtrados = df_cantidades["ID"].iloc[0:10]
IDs_filtrados
```

```python
df_datos[df_datos["ID"].isin(IDs_filtrados)]["ITEM"].unique()
```

```python
# grafico de los más vendidos sns matplotlib
```

```python

df_precio = df_columnas.groupby("ID").mean("precio_unitario")[["precio_unitario"]].sort_values("precio_unitario", ascending=False).reset_index()
#df_cantidades
df_precio
```

```python
IDs_filtrados_precio = df_precio["ID"].iloc[-110:-100]
IDs_filtrados_precio
```

```python
# Supongamos que tienes un DataFrame llamado df_datos con la información original
# y tienes una lista de IDs filtrados llamada IDs_filtrados_precio

# Paso 1: Filtra las filas con los IDs deseados y selecciona las columnas necesarias
df_menos = df_datos[df_datos["ID"].isin(IDs_filtrados_precio)][["ITEM", "precio_unitario", "CANTIDAD"]]

# Paso 2: Agrupa por "ITEM" y calcula el precio unitario promedio y la suma de la cantidad
resumen_df = df_menos.groupby("ITEM").agg({
    "precio_unitario": "mean",  # Calcula el promedio del precio unitario
    "CANTIDAD": "sum"           # Calcula la suma de la cantidad
}).reset_index()  # Reinicia el índice para obtener un DataFrame plano

# Visualiza el DataFrame resumen
resumen_df
```
